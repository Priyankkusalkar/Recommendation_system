#### Recommendation System:

Design a function to recommend anime based on cosine similarity.
Given a target anime, recommend a list of similar anime based on cosine similarity scores.
Experiment with different threshold values for similarity scores to adjust the recommendation list size.



```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MultiLabelBinarizer
from sklearn.metrics.pairwise import cosine_similarity

def preprocess_anime_data(anime_data):
    # Drop rows with missing values in 'rating', 'members', or 'genre'
    anime_data.dropna(subset=['rating', 'members', 'genre'], inplace=True)
    
    # Convert 'rating' and 'members' columns to numeric, forcing errors to NaN
    anime_data['rating'] = pd.to_numeric(anime_data['rating'], errors='coerce')
    anime_data['members'] = pd.to_numeric(anime_data['members'], errors='coerce')
    
    # Drop rows where 'rating' or 'members' are NaN after conversion
    anime_data.dropna(subset=['rating', 'members'], inplace=True)
    
    # Ensure 'rating' is float and 'members' is int
    anime_data['rating'] = anime_data['rating'].astype(float)
    anime_data['members'] = anime_data['members'].astype(int)
    
    # Debugging: Print the first few rows after conversion to check data types
    print("Data after cleaning and type conversion:")
    print(anime_data.head())
    print(anime_data.dtypes)
    
    # Convert genre column into one-hot encoded representation
    mlb = MultiLabelBinarizer()
    genre_encoded = pd.DataFrame(mlb.fit_transform(anime_data['genre'].str.split(', ')),
                                 columns=mlb.classes_,
                                 index=anime_data.index)
    
    # Concatenate one-hot encoded genres with original DataFrame
    anime_data_encoded = pd.concat([anime_data.drop(columns=['genre']), genre_encoded], axis=1)
    
    # Normalize numerical columns (rating and members)
    scaler = StandardScaler()
    anime_data_encoded[['rating', 'members']] = scaler.fit_transform(anime_data_encoded[['rating', 'members']])
    
    return anime_data_encoded

def recommend_similar_anime(target_anime, anime_data, threshold=0.5, top_n=10):
    # Check if target anime exists in the dataset
    if target_anime not in anime_data['name'].values:
        print("Target anime not found in the dataset")
        return []
    
    # Get the feature vector of the target anime
    target_index = anime_data[anime_data['name'] == target_anime].index[0]
    target_vector = anime_data.drop(columns=['anime_id', 'name']).iloc[target_index].values.reshape(1, -1)
    
    # Calculate cosine similarity between target anime and all other anime
    cosine_sim = cosine_similarity(anime_data.drop(columns=['anime_id', 'name']), target_vector).flatten()
    
    # Get indices of anime sorted by cosine similarity
    sim_indices = cosine_sim.argsort()[::-1]
    
    # Filter anime based on similarity threshold and remove the target anime
    similar_anime = [anime_data.iloc[i]['name'] for i in sim_indices if cosine_sim[i] >= threshold and anime_data.iloc[i]['name'] != target_anime]
    
    # Return top-N similar anime
    return similar_anime[:top_n]

# Load the dataset
anime_data = pd.read_csv(r"C:\Users\Laptop\Downloads\Recommendation System (1)\Recommendation System\anime.csv")

# Debugging: Print the first few rows to understand the structure and types of data
print("Initial data:")
print(anime_data.head())
print(anime_data.dtypes)

# Preprocess the anime dataset
anime_data_encoded = preprocess_anime_data(anime_data)

# Example usage:
target_anime = "Kimi no Na wa."
recommended_anime = recommend_similar_anime(target_anime, anime_data_encoded, threshold=0.7, top_n=5)
print("Recommended Anime:")
print(recommended_anime)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MultiLabelBinarizer
from sklearn.metrics.pairwise import cosine_similarity

def preprocess_anime_data(anime_data):
    # Drop rows with missing values
    anime_data.dropna(subset=['rating', 'members', 'genre'], inplace=True)
    
    # Convert 'rating' and 'members' columns to numeric, forcing errors to NaN
    anime_data['rating'] = pd.to_numeric(anime_data['rating'], errors='coerce')
    anime_data['members'] = pd.to_numeric(anime_data['members'], errors='coerce')
    
    # Drop rows where 'rating' or 'members' are NaN after conversion
    anime_data.dropna(subset=['rating', 'members'], inplace=True)
    
    # Ensure 'rating' is float and 'members' is int
    anime_data['rating'] = anime_data['rating'].astype(float)
    anime_data['members'] = anime_data['members'].astype(int)
    
    # Convert genre column into one-hot encoded representation
    mlb = MultiLabelBinarizer()
    genre_encoded = pd.DataFrame(mlb.fit_transform(anime_data['genre'].str.split(', ')),
                                 columns=mlb.classes_,
                                 index=anime_data.index)
    
    # Concatenate one-hot encoded genres with original DataFrame
    anime_data_encoded = pd.concat([anime_data.drop(columns=['genre']), genre_encoded], axis=1)
    
    # Normalize numerical columns (rating and members)
    scaler = StandardScaler()
    anime_data_encoded[['rating', 'members']] = scaler.fit_transform(anime_data_encoded[['rating', 'members']])
    
    return anime_data_encoded

def recommend_similar_anime(target_anime, anime_data, threshold=0.5, top_n=10):
    # Check if target anime exists in the dataset
    if target_anime not in anime_data['name'].values:
        print("Target anime not found in the dataset")
        return []
    
    # Get the feature vector of the target anime
    target_index = anime_data[anime_data['name'] == target_anime].index[0]
    target_vector = anime_data.drop(columns=['anime_id', 'name', 'type', 'episodes']).iloc[target_index].values.reshape(1, -1)
    
    # Calculate cosine similarity between target anime and all other anime
    cosine_sim = cosine_similarity(anime_data.drop(columns=['anime_id', 'name', 'type', 'episodes']), target_vector).flatten()
    
    # Get indices of anime sorted by cosine similarity
    sim_indices = cosine_sim.argsort()[::-1]
    
    # Filter anime based on similarity threshold and remove the target anime
    similar_anime = [anime_data.iloc[i]['name'] for i in sim_indices if cosine_sim[i] >= threshold and anime_data.iloc[i]['name'] != target_anime]
    
    # Return top-N similar anime
    return similar_anime[:top_n]

# Load the dataset
anime_data = pd.read_csv(r"C:\Users\Laptop\Downloads\Recommendation System (1)\Recommendation System\anime.csv")

# Preprocess the anime dataset
anime_data_encoded = preprocess_anime_data(anime_data)

# Debugging: Print the first few rows to understand the structure and types of data
print("Initial data:")
print(anime_data.head())
print(anime_data.dtypes)

# Example usage:
target_anime = "Kimi no Na wa."
recommended_anime = recommend_similar_anime(target_anime, anime_data_encoded, threshold=0.7, top_n=5)
print("Recommended Anime:")
print(recommended_anime)
```
